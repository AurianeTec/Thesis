```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
 
plt.rcParams['figure.figsize'] = (8, 8)
```

```python
IRIS_GPM_clean = gpd.read_file("data/geographical_data/IRIS_code_geometry.geojson")
```

```python
# IRIS-level income data
# from https://www.insee.fr/fr/statistiques/6049648#consulter

revenu_decl_IRIS_2019 = pd.read_csv("data/IRIS_DATA/BASE_TD_FILO_DEC_IRIS_2019.csv")
merged = revenu_decl_IRIS_2019.merge(IRIS_GPM_clean, left_on='IRIS', right_on='CODE_IRIS', how='right')
merged = gpd.GeoDataFrame(merged, geometry = 'geometry').to_crs(epsg = 4326)#, allow_override = True)
merged.plot('DEC_EQ19', alpha=0.5, linewidth=0.5)
#GPM.plot(ax=ax, alpha=0.5, edgecolor = 'black', linewidth = 2, facecolor = 'none')
```

```python
# IRIS-level income data (META DATA)
meta_data = pd.read_csv("data/IRIS_DATA/meta_BASE_TD_FILO_DISP_IRIS_2019.csv", sep=';')
meta_data.head()

merged = meta_data.merge(IRIS_GPM_clean, how='right',left_on='COD_MOD', right_on='CODE_IRIS')
merged = gpd.GeoDataFrame(merged, geometry = 'geometry').to_crs(epsg = 4326)
merged.plot('CODE_IRIS', alpha=0.5, linewidth=0.5)
#GPM.plot(ax=ax, alpha=0.5, edgecolor = 'black', linewidth = 2, facecolor = 'none')
```

```python
# IRIS-level pop data
# from https://www.insee.fr/fr/statistiques/6543200#consulter

pop_IRIS_2019 = pd.read_csv("data/struct-pop-2019/struct-pop-2019.csv", delimiter=';', low_memory=False)
merged = pop_IRIS_2019.merge(IRIS_GPM_clean, left_on='IRIS', right_on='CODE_IRIS', how='right')
merged = gpd.GeoDataFrame(merged, geometry = 'geometry').to_crs(epsg = 4326)#, allow_override = True)
merged.plot('P19_POP0019', alpha=0.5)
#GPM.plot(ax=ax, alpha=0.5, edgecolor = 'black', linewidth = 2, facecolor = 'none')
```
